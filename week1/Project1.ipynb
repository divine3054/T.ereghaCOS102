```python
def simple_interest(principal, rate, time):
    A=principal*(1+(rate/100)*time)
    return A

principal=int(input("enter principal"))  
rate=int(input("enter rate")) 
time=int(input("enter time")) 

    
simple_interest_answer=simple_interest(principal,rate,time) 
print(simple_interest_answer)
```

```python
def compound_interest(principal, rate, time,compound_per_year):
    A=principal*(1+(rate/compound_per_year))**(compound_per_year*time)
    return A
principal=int(input("enter principal"))  
rate=int(input("enter rate")) 
time=int(input("enter time")) 
compound_per_year=int(input("enter compound per year")) 

    
compound_interest_answer=compound_interest(principal,rate,time,compound_per_year) 
print(compound_interest_answer)
```
