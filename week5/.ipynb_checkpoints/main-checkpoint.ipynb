```python
#practice1 | Cell 1 ss
class SST:
    

    def __init__(self, name):
        self.name = name
        self.prog1 = "Computer Science"
    def speak(self):
        print("\nMy name is",self.name)
        print("I study",self.prog1)

stud1= SST(input("Enter student1 name: "))
stud2= SST(input("Enter student2 name: "))

stud1.speak()
stud2.speak()
```

```python
class Addition:
    def __init__(self, f, s):
        self.first = f
        self.second = s
        self.answer=None
    def calculate(self):
        self.answer = self.first + self.second
    
    def display(self):
        print("\nFirst number:",self.first)
        print("Second number:",self.second)
        print("Sum:",self.answer)

    num1 = int(input("\nEnter first number: "))
    num2 = int(input("Enter second number: "))

    obj = Addition(num1, num2)
    obj.calculate()    
    obj.display()   
    
```
