```python
import tkinter as tk
from tkinter import messagebox
import random

# Data
EMPLOYEES = [
    "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu",
    "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo",
    "Abraham Jones", "Nicole Anide", "Kosi Korso", "Adele Martins",
    "Emmanuel Ojo", "Ajayi Fatima"
]

TASKS = [
    "Loading", "Transporting", "Reviewing Orders",
    "Customer Service", "Delivering Items"
]

# OOP Class
class Employee:
    def __init__(self, name):
        self.name = name

    def check_employee(self):
        return self.name in EMPLOYEES

    def take_attendance(self):
        return f"Attendance taken for {self.name}"

    def assign_task(self):
        task = random.choice(TASKS)
        return f"Task assigned: {task}"

    def refuse_access(self):
        return "Access Denied! You are not an employee."


# GUI App
class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Employee Management System")
       

        # Label
        self.label = tk.Label(root, text="Enter Employee Name:")
        self.label.pack(pady=10)

        # Entry
        self.entry = tk.Entry(root, width=30)
        self.entry.pack(pady=5,m)

        # Button
        self.button = tk.Button(root, text="Submit", command=self.process)
        self.button.pack(pady=10)

        # Output
        self.output = tk.Label(root, text="", wraplength=350, justify="left")
        self.output.pack(pady=20)

    def process(self):
        name = self.entry.get().strip()
        emp = Employee(name)

        if emp.check_employee():
            attendance = emp.take_attendance()
            task = emp.assign_task()
            result = f"Welcome {name}!\n{attendance}\n{task}"
        else:
            result = emp.refuse_access()

        self.output.config(text=result)


# Run App

root = tk.Tk()
app = App(root)
root.mainloop()
```
